```python
import binascii

def parse_midi(name):
    midi_file = f'./{name}.mid'

    with open(midi_file, 'rb') as f:
        hexdata = binascii.hexlify(f.read())
    hexdata_str = hexdata.decode("utf-8")
    
    MTrk = '4d54726b'
    arr = []
    for i,chunk in enumerate(hexdata_str.split(MTrk)):
        if i == 0:
            arr.append(chunk)
        else:
            arr.append (MTrk + chunk)

    return arr
```

```python
new = parse_midi(name='test_new/beethoven_from_testnet')
old = parse_midi(name='test/beethoven_test')
```
